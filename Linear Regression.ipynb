```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

#Sample data
x = np.linspace(0, 10, 100)
y = 2 * x + 5 + np.random.randn(100) * 2

#Create a Data frame
df=pd.DataFrame({'x': x, 'y': y})

#Fit Linear Regression Model
X = sm.add_constant(df['x'])
model = sm.OLS(df['y'], X).fit()

#Plot the data and regression line
plt.scatter(df['x'], df['y'])
plt.plot(df['x'], model.fittedvalues, color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#Sample data
np.random.seed(0)
x = np.linspace(-3, 3, 100)
y = x**2+np.random.rand(100)*2

#create polynomial features
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(x.reshape(-1,1))

#fit the lin reg model
model=LinearRegression().fit(X_poly, y)

#predict values
y_pred = model.predict(X_poly)

#Plot the data and reg curve
plt.scatter(x, y, label='data')
plt.plot(x, y_pred, color='black')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

#fit the lin reg model
model = LinearRegression().fit(X,y)

#Make predictions
y_pred = model.predict(X)

#Plot the data and regression line
plt.scatter(X,y)
plt.plot(X, y_pred, y - y_pred)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('residual plot')
plt.axhline(y = 0, color = 'purple', 
linestyle = '--')
plt.show()

#Histogram of residuals
plt.hist(y - y_pred, bins=20)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```
